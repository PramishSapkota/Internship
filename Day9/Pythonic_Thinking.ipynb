# **Iterator**

An iterator is an object that contains a countable number of values

"iter()" and "next()"

**Scenario:**
Create a tuple called mytuple with the values "apple", "banana", "cherry".Create a variable myit and use iter() on the tuple
Print the first item using next()

```python
# Create a tuple
mytuple = ("apple", "banana", "cherry")

# Create an iterator
myit = iter(mytuple)

# Print the first item
print(next(myit))
print(next(myit))
```

To create an object/class as an iterator  we use the methods \_\_iter\_\_() and \_\_next\_\_()

```python
class MyNumbers:
  def __iter__(self):
    self.a = 1
    return self

  def __next__(self):
    if self.a <= 20:
      x = self.a
      self.a += 1
      return x
    else:
      raise StopIteration

myclass = MyNumbers()
myiter = iter(myclass)
print(next(myiter))

# for x in myiter:
#   print(x)
```

# **Generator**

Python generators are created similar to function with *def* keyword and *yield* statement instead of return

```python
def a_generator():
  yield "1st iteration"
  yield "2nd iteration"

# create a generator object from the generator function by calling the function
# like we would any other function
a = a_generator()
print(a)
print(type(a))#<class 'generator'>

x = next(a)
print(type(x),"\n")
print(x)

print(next(a))
# print(next(a)) # throws an error StopIteration if it doesnt get any yield value
```

## send() method
allows you to send a value to the generator:

```python
def echo_generator():
  while True:
    received = yield
    print("Received:", received)

gen = echo_generator()
next(gen) # Prime the generator
# if we dont prime we will get an TypeError
# TypeError:can't send non-None value to a just-started generator

gen.send("Hello")
gen.send("World")
```

# close() Method
The close() method stops the generator:

```python
def my_gen():
  try:
    yield 1
    yield 2
    yield 3
  finally:
    print("Generator closed")

gen = my_gen()
print(next(gen))
gen.close()
```

```python
def my_generator(n):

    # initialize counter
    value = 0

    # loop until counter is less than n
    while value < n:

        # produce the current value of the counter
        yield value

        # increment the counter
        value += 1

# iterate over the generator object produced by my_generator
for value in my_generator(3):

    # print each value produced by generator
    print(value)

```

```python
generator = my_generator(3)

print(next(generator))  # 0
print(next(generator))  # 1
print(next(generator))  # 2
# print(next(generator))  # throws an error StopIteration
```

```python
import functools
import time

@functools.lru_cache()
def my_generator(n):

    # initialize counter
    value = 0

    # loop until counter is less than n
    while value < n:

        # produce the current value of the counter
        yield value

        # increment the counter
        value += 1

# iterate over the generator object produced by my_generator
start = time.time()
for value in my_generator(3):

    # print each value produced by generator
    print(value)

print(f"{time.time() - start:.2f} seconds\n")

print("Again...")
# iterate over the generator object produced by my_generator
start = time.time()
# Second iteration - the cached generator is already exhausted!
gen2 = my_generator(3)
print(list(gen2))  # [] - EMPTY!
print(f"{time.time() - start:.2f} seconds\n")


start = time.time()
for value in my_generator(4):

    # print each value produced by generator
    print(value)

print(f"{time.time() - start:.2f} seconds\n")
```

my takeaway is that generator do the same thing as iterator but with less code

## why use generator?


1. Generator can do the same thing with less code.
2. **Memory Efficient:**
 Functions will create the entire sequence to return in memory before returning the result. While generator only produces one item at a time and hence is memory friendly.


**3. Pipelining Generators**

Multiple generators can be used to pipeline a series of operations.

```python
def fibonacci_numbers(nums):
    x, y = 0, 1
    for _ in range(nums):
        x, y = y, x+y
        yield x

def square(nums):
    for num in nums:
        yield num**2

print(sum(square(fibonacci_numbers(10))))

# Output: 4895
```

# **Zip**

It is used to combine two or more iterables (like lists, tuples, strings, dictionaries, etc.) into a single iterator of tuples.

* No iterable -> empty list.
* One iterable -> each element becomes a one-item tuple.
* Two iterables -> pairs elements at matching indexes.

```python
a = [1, 2, 3]
b = ['a', 'b', 'c']

# # No iterable are passed
# print("No Iterable")
# res = zip()
# print(res)
# print(type(res))
# print(list(res))
# print()

# # One iterable is passed
# print("1 Iterable")
# res = zip(a)
# print(res)
# print(type(res))
# print(list(res))
# print()

# Two iterables are passed
# print("2 Iterable")
res = zip(a, b)
# print(res)
# print(type(res))
# print(list(res))
# # the zip are very short lived can only use them once
# print(list(res)) #o/p: []
# print()

# # but if we make list like below we can use as many times we want
# res = list(zip(a, b))
# print(res)
# print(res)

for i,j in res:
  print(i,j)

```

if iterables of different Lengths are passed pairing stops as soon as the shortest iterable runs out of elements.

```python
names = ['Ram', 'Shyam', 'Hari']
scores = [88, 94]
# hari is left out here
res = zip(names, scores)
print(list(res))
```

# map()

## **Syntax**

map(function, iterables)

there can be multiple iterables

```python
def myfunc(a, b):
  return a + b

x = map(myfunc, ('apple', 'banana', 'cherry'), ('orange', 'lemon', 'pineapple'))

print(x)

#convert the map into a list, for readability:
print(list(x))
```

# filter()

The filter() function selects elements from an iterable based on the result of a function.

## Syntax:

filter(function, iterable)

altough the syntax is similar to map, the difference is that the **function in filter must return True or False**. Also another thing is that **filter has only one iterable**.  

```python
numbers = [1, 2, 3, 4, 5, 6]

evens = filter(lambda x: x % 2 == 0, numbers)

print(list(evens))
```

```python
def is_even(a):
  return a % 2 == 0

numbers = [1, 2, 3, 4, 5, 6]

evens = filter(is_even, numbers)

print(list(evens))
```

# enumerate function

```python
fruits = ['apple', 'banana', 'cherry']

# Basic usage (starts at 0)
for index, fruit in enumerate(fruits):
    print(f"Index {index}: {fruit}")

# Custom start index (starts at 1)
for index, fruit in enumerate(fruits, start=1):
    print(f"Item {index}: {fruit}")
```

# async

* **async** keyword is used before a function to make it a asynchronous function.
* **await** keyword is used to execute the asynchronous function
* **asyncio.gather** is used to run the functions simultaneously.

```python
import asyncio

async def task1():
    print("Task 1 started")
    await asyncio.sleep(3)
    print("Task 1 finished")

async def task2():
    print("Task 2 started")
    await asyncio.sleep(1)
    print("Task 2 finished")

async def main():
    await asyncio.gather(task1(), task2())  # Runs both tasks together

# await asyncio.run(main())
# ''' Above line causes error in colab because Colab already has
# an active asyncio event loop. asyncio.run() tries to start a new one,
# which conflicts with the existing one.
# '''

await main() # This is the fix for Colab
```
