#**Class And Objects**

The self parameter is a reference to the current instance of the class.It is used to access properties and methods that belong to the class.

```python
class Car:
  def __init__(self, color, brand):
    self.color = color
    self.brand = brand

# object creation
c1 = Car("Red", "Tata")

print(f"There is a {c1.color} colored {c1.brand} car")
```

# **Class Properties & Object Properties**
Properties defined inside __init__() belong to each object (instance properties).Properties defined outside methods belong to the class itself (class properties) and are shared by all objects:

```python
class Person:
  species = "Human" # Class property

  def __init__(self, name):
    self.name = name # Instance property

p1 = Person("Hari")
p2 = Person("Ram")

print(p1.name)
print(p2.name)
print(p1.species)
print(p2.species)
```

Class properties are changed using class name

While object properties can be changed using the object

```python
class Person:
  lastname = "Shrestha"

  def __init__(self, name):
    self.name = name

p1 = Person("Hari")
p2 = Person("Ram")

print(f"Full name for p1: {p1.name} {p1.lastname}")
print(f"Full name for p2: {p2.name} {p2.lastname}")

#this way all objects lastname are changed
Person.lastname = "Sharma"

print(p1.lastname)
print(p2.lastname)

p1.name = "Pramish"
p1.lastname = "Sapkota"
print(f"\nFull name for p1: {p1.name} {p1.lastname}")
```

# **Inheritance**

To create a child class the parent class as a parameter when creating the child class:

##**Single Inheritance**

```python
class Car:
  color = 'black'

# @ staticmethod is used when we dont need self parameter in a method
  @staticmethod
  def car_start():
    print('Car Started')

  @staticmethod
  def car_stop():
    print('Car Stopped')


class Ford(Car): #single inheritance
  def __init__(self, name, brand):
    self.name = name
    self.brand = brand

f1=Ford('Mustang', "electric")
print(f1.name)
print(f1.brand)

# Parent class properties
print(f1.color)
f1.car_start()
f1.car_stop()

```

## **Multiple Inheritance**

```python
class Mammal:
    def mammal_info(self):
        print("Mammals can give direct birth.")

class WingedAnimal:
    def winged_animal_info(self):
        print("Winged animals can flap.")

class Bat(Mammal, WingedAnimal):
    pass

# create an object of Bat class
b1 = Bat()

# bat class can access methods from Mammal & WingedAnimal
b1.mammal_info()
b1.winged_animal_info()
```

###MRO
Python uses Method Resolution Order(MRO) which solves the diamond problem occuring in other languages like c++. The MRO ensures that methods are resolved in a predictable and consistent order, eliminating ambiguity and duplication.

```python
class Shape:
    def draw(self):
        print("class shape")

class Circle(Shape):
    def draw(self):
        print("class circle")

class Square(Shape):
    def draw(self):
        print("class square")

class Cylinder(Circle, Square):
    pass

obj = Cylinder()
obj.draw()  # Resolves using the MRO
print(Cylinder.mro())  # Displays the MRO
# cylinder -> circle -> square -> shape
```

###**super()**

super method ensures that the execution is in predictable order and works alongside MRO. It is called cooperative multiple inheritance

```python
class shape:
    def greet(self):
        print("class shape called")

class circle(shape):
    def greet(self):
        super().greet()
        print("class circle called")

class square(shape):
    def greet(self):
        super().greet()
        print("class square called")

class cylinder(circle, square):
    def greet(self):
        super().greet()
        print("class cylinder called")

d = cylinder()
d.greet() # cooperative multiple inheritance
```

#**Polymorphism**

if parent class and child class have same function name then child class fn overrides parent fn.



```python
class Vehicle:
  @staticmethod
  def move():
    print("Vehicle moves on road")

class Car(Vehicle):
  @staticmethod
  def move():
    print("Car moves on road")

nexon= Car()
nexon.move()
```

to also use parent class functions we need to use super() function

```python
class Person:
  def __init__(self, fname, lname):
    self.firstname = fname
    self.lastname = lname

  def printname(self):
    print(self.firstname, self.lastname)

class Student(Person):
  def __init__(self, fname, lname, year):
    super().__init__(fname, lname)# parent class function is also being used
    self.graduationyear = year

x = Student("Pramish", "Sapkota", 2026)
print(f"{x.firstname} {x.lastname} is graduating on {x.graduationyear}")
```

#**Encapsulation**

Encapsulation refers to the bundling of attributes and methods inside a single class.

It prevents outer classes from accessing and changing attributes and methods of a class.

 "\_attribute" or fn  for protected (single underscore)

"\__attribute" or fn for private (double underscore)





```python
class Computer:

    def __init__(self):
        self.__maxprice = 900   # private variable

    def sell(self):
        print("Selling Price:", self.__maxprice)

    def setMaxPrice(self, price):
        self.__maxprice = price


c = Computer()
c.sell()

# trying to change directly
c.__maxprice = 1000
c.sell()

# using setter function
c.setMaxPrice(1000)
c.sell()
```

#**Operator Overload**

**<**

object.__lt__(self, other)

**<=**

object.__le__(self, other)

**=**

object.__eq__(self, other)

**!=**

object.__ne__(self, other)

**>**

object.__gt__(self, other)

**>=**

object.__ge__(self, other)

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    # overload < operator
    def __lt__(self, other):
        return self.age < other.age

p1 = Person("Alice", 20)
p2 = Person("Bob", 30)

print(p1 < p2)  # prints True
print(p2 < p1)  # prints False
```
