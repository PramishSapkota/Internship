**collection module**

| Built-in Type  | Problem                          | `collections` Solution                               |
| -------------- | -------------------------------- | ---------------------------------------------------- |
| `dict`         | Missing keys, counting, ordering | `defaultdict`, `Counter`, `OrderedDict`              |
| `list`         | Slow queue operations            | `deque`                                              |
| `tuple`        | No field names                   | `namedtuple`                                         |

# **NamedTuple**

it is like a mix of dictionary and class. But can also access values using index which dictionary lack.

acts as class bcz its type is 'type' and for the instance its type is also similar to the object.

```python
#to see type of class and object

class X:
    pass

x = X()
print("Type of class: ",type(X))
print("Type of object: ",type(x))
```

```python
# Python code to demonstrate namedtuple()
from collections import namedtuple

# Declaring namedtuple()
Student = namedtuple('Student', ['name', 'age', 'DOB'])
print(f"Type of Student: {type(Student)}")
# Adding values
S = Student('Ram', '31', '19990121')
print(S)
print(f"Type of S: {type(S)}\n")

# Access using index
print("The Student age using index is : ", end="")
print(S[1])

# Access using name
print("The Student name using keyname is : ", end="")
print(S.name)

# Access using getattr()
print("The Student DOB using getattr() is : ", end="")
print(getattr(S, 'DOB'))
```

## Conversion Operations

* Using _make()
* Using _asdict()
* Using “**” (double star) operator

```python
from collections import namedtuple

# Declaring namedtuple()
Student = namedtuple('Student',
                                 ['name', 'age', 'DOB'])

# Adding values
S1 = Student('Ram', '31', '19990121')
print(f"Type of S1: {type(S1)}")
print(S1)

# initializing iterable
li = ['Manoj', '19', '411997']

di = {'name': "Nikhil", 'age': 19, 'DOB': '1391997'}

# using _make() to return namedtuple()
print("\nThe namedtuple instance using _make() is  : ")
S2 = Student._make(li)
print(f"Type of S2: {type(S2)}")
print(S2)


S3 = S1._asdict()
'''
_asdict() converts into dictionary it doesnt convert the dictionary
into nametuple for that we have "**" (double star) operator
'''
print("\nThe OrderedDict instance using namedtuple is  : ")
print(f"Type of S3: {type(S3)}")
print(S3)

# using ** operator to return namedtuple from dictionary
print("The namedtuple instance from dict is  : ")
S4 = Student(**di)
print(f"Type of S4: {type(S4)}")
print(S4)
```

```python
SS = Student._make(di)
print(f"Type of SS: {type(SS)}")
print(SS)
```

```python
S3 = S1._asdict()
S3['name'] = 'sita'
print(S3)
print(S1)

S1._asdict()['name'] = 'sita'
print(S1)
```

## _replace()

```python
import collections

# Declaring namedtuple()
Student = collections.namedtuple('Student',
                           ['name', 'age', 'DOB'])

# Adding values
S = Student('Nandini', '19', '2541997')

# ._replace returns a new namedtuple,
# it does not modify the original
print("returns a new namedtuple : ")
print(S._replace(name='Manjeet')) # this didnt replace in original S
print(S)
S = S._replace(name='Manjeet')
print(S)
```

# **Counter**

A counter is a sub-class of the dictionary. It is used to keep the count of the elements in an iterable in the form of an unordered dictionary where the key represents element in the iterable and value represents the count of that element in the iterable.

```python
from collections import Counter

text = "cbanana"
freq = Counter(text)

print(freq)
print(list(freq.elements()))
#['c', 'b', 'a', 'a', 'a', 'n', 'n']
# c came 1st then n then a then n
```

```python
from collections import Counter

print(Counter())                           # a new, empty counter
count = Counter('gallahad')                # a new counter from an iterable
print(count,"\n")

#dictionary methods                # a new counter from an iterable
print(count.items())                # a new counter from an iterable
print(count.keys())                # a new counter from an iterable
print(count.values())                # a new counter from an iterable
```

```python
from collections import Counter

print(Counter({'red': 4, 'blue': 2}))     # a new counter from a mapping
print(Counter(cats=4, dogs=8))           # a new counter from keyword args

nums = [1,1,1,2,2,3]

c = Counter(nums)
print(c)
print(c.most_common(2))# gives list of tuples
```

# **deque**

 It is a double-ended queue.

Fast insertion/removal from both ends.

```python
from collections import deque

q = deque()
print(q)

#adding items
q.append(10)
q.append(20)
print(q)
q.appendleft("a")
q.appendleft(True)
print(q)

#removing items
print(q.pop()) # removes  the rightmost element(last)
print(q.popleft())# removes  the leftmost element(1st)
print(q)
```

# **ChainMap**

Takes many dictionaries and returns a list of dictionaries.

```python
from collections import ChainMap

d1 = {'a': 1, 'b': 2}
d2 = {'c': 3, 'd': 4}
d3 = {'a': 5, 'f': 6}

# Defining the chainmap
c = ChainMap(d1, d2, d3)
print(type(c))
print(c,"\n")
# it prints the object

# printing chainMap using maps
print(c.maps)
#need to use maps to treat the Chainmap as list of dictionaries
print(c.maps[0],"\n")

print(c.keys())
print(list(c.keys()),"\n")

print(c.values())
print(list(c.values()))
```

## reversed() and new_child()

```python
import collections

# initializing dictionaries
dic1 = {'a': 1, 'b': 2}
dic2 = {'b': 3, 'c': 4}
dic3 = {'f': 5}

# initializing ChainMap
chain = collections.ChainMap(dic1, dic2)

print("Original ChainMap:")
print(chain.maps)

# add new child
chain1 = chain.new_child(dic3)

print("\nAfter new_child():")
print(chain1.maps)

# value before reversing
print("\nValue of b before reversing:", chain1['b'])

# reverse the maps properly
reversed_maps = list(reversed(chain1.maps))

print("\nReversed maps:")
print(reversed_maps)

# assign reversed list back
chain1.maps = reversed_maps

# value after reversing
print("\nValue of b after reversing:", chain1['b'])

# print dictionaries one by one
print("\nPrinting dictionaries after reversing:")
for d in chain1.maps:
    print(d)
```

# **OrderedDict**

```python
from collections import OrderedDict

d = OrderedDict()

d["a"] = 1
d["b"] = 2
d["c"] = 3

d.move_to_end("a")

print(d)

#popping from left
d2 = OrderedDict([('a', 1), ('b', 2), ('c', 3)])
d3 = OrderedDict([('a', 1), ('b', 2), ('c', 3)])
# d3 = dict([('a', 1), ('b', 2), ('c', 3)])
print(d2)
print(d3)
res = d3.popitem(last=False)  # Remove firstinserted item
res1= d2.popitem()
print(res)
print(res1)

#Checking equality
od1 = OrderedDict([('a', 1), ('b', 2), ('c', 3)])
od2 = OrderedDict([('c', 3), ('b', 2), ('a', 1)])
print(od1 == od2)

d1 = {'a': 1, 'b': 2, 'c': 3}
d = {'c': 3, 'b': 2, 'a': 1}
print(d1==d)
```

# **UserDict**

Wrapper around dictionaries for easier subclassing

```python
from collections import UserDict

class MyDict(dict):
    def __setitem__(self, key, value):
        super().__setitem__(key.upper(), value)#here we are saving the keys in uppercase

      # Prevents using 'del' on dictionary
    def __del__(self):
        print("Deletion isn't allowed")

d = MyDict(name = 'Ram')
print(d)

e = MyDict()
e["name"] = "Bob"

print(e)
del e

class MyDict1(UserDict):
    def __setitem__(self, key, value):
        super().__setitem__(key.upper(), value)#here we are saving the keys in uppercase

      # Prevents using 'del' on dictionary
    def __del__(self):
        print("Deletion isn't allowed")

d = MyDict1(naME = 'Ram')
print(d)

e = MyDict1()
e["namE"] = "Bob"

print(e)
del e
```

# **UserList**

```python
from collections import UserList

# Creating a list where deletion is not allowed
class MyList(UserList):

    # Prevents using remove() on list
    def remove(self, s=None):
        print("Deletion not allowed")

    # Prevents using pop() on list
    def pop(self, s=None):
        print("Deletion not allowed")

    def custom_fn(self):
      print("This is my custom function")

# Create an instance of MyList
l = MyList([1, 2, 3, 4])
print("Original List")

# Append 5 to the list
l.append(5)
print("After Insertion")
print(l)

l.custom_fn()

# Attempt to remove an item (will raise error)
l.remove()
```

# **UserString**

```python
from collections import UserString

# Creating a Mutable String
class Mystring(UserString):

    # Function to append to string
    def append(self, s):
        self.data += s

    # Function to remove from string
    def remove(self, s):
        self.data = self.data.replace(s, "")

    def custom_fn(self):
      print("This is my custom function")

# Driver's code
s1 = Mystring("Geeks")
print("Original String:", s1.data)

# Appending to string
s1.append("s")
print("String After Appending:", s1.data)

# Removing from string
s1.remove("e")
print("String after Removing:", s1.data)

s1.custom_fn()
```
