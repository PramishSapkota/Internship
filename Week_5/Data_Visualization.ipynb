# matplotlib

```python
import matplotlib.pyplot as plt

x = (1, 2, 3, 4, 5)
y = (1, 4, 9, 16, 25)

plt.plot(x, y, linewidth=2, marker='o', color='green')
plt.title("Square Numbers", fontsize=24)
plt.xlabel("Value",fontsize=14)
plt.ylabel("Square of Value", fontsize=14)
plt.grid(False)
# plt.show()
# plt.savefig('squares_plot.png')
```

we can use the keyword argument `markersize` or the shorter version `ms` to set the size of the markers.

Similarly, `markeredgecolor `as `mec`, `markerfacecolor` as `mfc`

also we can use `linestyle` as `ls`
 and `':'` or `'dotted'`,
 and `'--'` or `'dashed'`,
 and `'-.'` or `	'dashdot'` line.

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0, 6])
ypoints = np.array([0, 250])

plt.subplot(2, 2, 1)
plt.plot(xpoints, ypoints,'o-r')
plt.title("Red Smooth Line", fontsize = 10)
# plt.show()

plt.subplot(2, 2, 2)
plt.plot(xpoints, ypoints,'X:g')
plt.title("Green Dotted Line",fontsize = 10)
# plt.show()

plt.subplot(2, 2, 3)
plt.plot(xpoints, ypoints,'p--b')
plt.title("Blue Dashed Line", fontsize=10)
# plt.show()

plt.subplot(2, 2, 4)
plt.plot(xpoints, ypoints,'D-.y')
plt.title("Yellow Dash Dot Line", fontsize = 10)
# plt.show()

# for manual adjustment
plt.subplots_adjust(hspace=0.5)

# for auto adjustment
# plt.tight_layout()
plt.show()
```

## Scatter Plot

```python
import matplotlib.pyplot as plt
import numpy as np

#day one, the age and speed of 13 cars:
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y, color='blue', label='Day 1')


#day two, the age and speed of 15 cars:
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])


plt.scatter(x, y, color='red', label='Day 2')

# Legend
plt.legend(bbox_to_anchor = (1,1))

plt.show()
```

## Bar Graph

default height and width is `0.8`

```python
import matplotlib.pyplot as plt
import numpy as np

arr1 = np.array(('Sun', 'Mon', 'Tue', 'Wed'))
arr2 = np.array((30,25,45,20))

plt.bar(arr1,arr2,width = 0.5)
plt.xlabel("Day")
plt.ylabel("Temp")
plt.title("Bar Graph for Day V/S Tmp")
plt.show()
```

### horizontal bar graph

```python
import matplotlib.pyplot as plt
import numpy as np

arr1 = np.array(('Sun', 'Mon', 'Tue', 'Wed'))
arr2 = np.array((30,25,45,20))

plt.barh(arr1,arr2,color = 'DarkGoldenRod', height = 0.5)
plt.ylabel("Day")
plt.xlabel("Temp")
plt.title("Bar Graph for Day V/S Tmp")
plt.show()
```

## Bar Graph

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = np.array([0.2, 0.1, 0, 0])
my_color = np.array(['c','m','y','r'])

plt.pie(y, labels = mylabels, explode = myexplode, shadow = True, colors= my_color)#, startangle= 180)
plt.show()
```

# seaborn

It is basically a wrapper around matplotlib

## histplot

```python
import seaborn as sns
import matplotlib.pyplot as plt

marks = [40, 45, 50, 55, 60, 62, 65, 70, 70, 75, 80]

sns.histplot(marks)

plt.show()

# plt.hist(marks,color='aqua')
# plt.show()
```

## scatterplot

```python
study_hours = [1,2,3,4,5]
marks = [40,50,60,70,80]

sns.scatterplot(
    x=study_hours,
    y=marks
)
```

## lineplot

```python
days = [1,2,3,4,5]
temp = [25,27,30,29,32]

sns.lineplot(x=days,y=temp)
```

## barplot

```python
subjects = ["Math","Physics","Chemistry"]
marks = [80,90,75]

sns.barplot(x=subjects,y=marks)
```

## boxplot

```python
marks = [40,45,50,55,60,65,70,150]
sns.boxplot(x=marks)
```

## heatmap

```python
import numpy as np

matrix = np.array([
    [1,2,3],
    [4,5,6],
    [7,18,19]
])

sns.heatmap(matrix,annot=True)
```
